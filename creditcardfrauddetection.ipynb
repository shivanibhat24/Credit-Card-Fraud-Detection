```python
#importing all Python Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a Pandas Dataframe
credit_card_data=pd.read_csv('creditcard.csv')
credit_card_data.head()
```

```python
credit_card_data.info()
```

```python
#Checking the number of missing values
credit_card_data.isnull().sum()
```

```python
#Let's check the distribution of legit and fraud transactions
credit_card_data['Class'].value_counts()
#0 indicates legit and 1 is for frauds....This dataset is unbalaced
```

```python
#Let's separate the data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
#Statistical Measures of Data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Comparing the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
#Undersampling the Data points
legit_sample=legit.sample(n=492) #Random sampling out of 2.8 lakh+ values
#Creating a sample dataset
new_dataset=pd. concat([legit_sample,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Splitting the Data into features and targets
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
print(X)
```

```python
#Splitting the data into training data and testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
print(X.shape,X_test.shape,X_test.shape)
print(Y.shape, Y_test.shape,Y_train.shape)
```

```python
#Model training
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
#Accuracy Score
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print("Accuracy on Training Data: ",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print("Accuracy on Test Data: ",test_data_accuracy)
```
